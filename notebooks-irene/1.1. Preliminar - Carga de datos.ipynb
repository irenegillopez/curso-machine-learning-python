# 3. Funci√≥n open

Para cargar datasets MUY pesados (>15GB)
- <b>Ventaja:</b> m√°s √≥ptimo que pandas ya que no carga el dataset de golpe, si no l√≠nea a l√≠nea
- <b>Inconveniente:</b> tienes que crear manualmente la tabla.

```python
filename4 = "customer-churn-model/Customer Churn Model.txt" # Fichero a abrir 
fullpath4 = mainpath + "/" + filename4 # Ruta completa
```

<b>Cuidado:</b> si abres el archivo en modo escritura (w) podr√≠as borrar su contenido

```python
df4 = open(fullpath4, 'r') # Abrimos el fichero en modo lectura (r)
```

Recorremos cada l√≠nea del dataset para convertirlo en un String con:

- next -> para leer de l√≠nea en l√≠nea
- strip -> quita espacios, reconoce tabuladores y saltos de l√≠nea (ej: \t o \n)
- split -> usa el delimitador (,) para separar los valores obtenidos

<b>CONCLUSI√ìN:</b> La 1¬™ l√≠nea obtenida corresponde a los nombres de las columnas -> por ende, el n¬∫ que obtengamos de valores en esa l√≠nea, es igual al n¬∫ de columnas.

```python
cols = next(df4).strip().split(",") #valor por l√≠nea
n_cols = len(cols) #n√∫mero de valores por l√≠nea
cols #valor por l√≠nea
```

```python
n_cols #n√∫mero de valores por l√≠nea
```

```python
#Construimos manualmente la tabla

counter = 0
main_dict = {}  #Creamos un diccionario = par-valor para asignar un valor a su par en cada columna.
for col in cols:  #Para cada columna de cada uno de los 21 valores x l√≠nea que hemos obtenido
    main_dict[col]=[]  #A√±√°dele un hueco
    
for line in df4:  #Para cada l√≠nea del dataset 
    values = line.strip().split(",")  #Divide el string en valores (values) con strip y split.
    for i in range (n_cols):  #Para cada columna que corresponda a esa l√≠nea 
        main_dict[cols[i]].append(values[i])  #As√≠gnale el valor de esa columnaxl√≠nea -> ej: 3x3 = Blanco
    counter += 1

print("El Dataset tiene %d filas y %d columnas" %(counter,n_cols))

df5 = pd.DataFrame(main_dict)  #La funci√≥n DataFrame crea una tabla con los valores del diccionario
df5.head()
```

### 2. Empleando la librer√≠a - "urllib3" y construyendo manualmente la tabla

```python
#primero importamos la librer√≠a y hacemos la conexi√≥n con la web de los datos
import urllib3
http = urllib3.PoolManager()
r = http.request('GET', medals_url)
print("El estado de la respuesta es %d" %(r.status))
response = r.data
    
#El objeto reponse contiene un string binario, as√≠ que lo convertimos a un string descodific√°ndolo en UTF-8
str_data = response.decode('utf-8')

#Dividimos el string en un array de filas, separ√°ndolo por intros
lines = str_data.split('\n')

#La primera l√≠nea contiene la cabecera, as√≠ que la extraemos
col_names = lines[0].split(",")
n_cols = len(col_names)

#Generamos un diccionario vac√≠o donde ir√° la informaci√≥n procesada desde la URL externa
counter = 0
main_dict = {}
for col in col_names:
    main_dict[col] = []

#Procesamos fila a fila la informaci√≥n para ir rellenando el diccionario con los datos como hicimos antes
for line in lines:
    #Nos saltamos la primera l√≠nea que es la que contiene la cabecera y ya tenemos procesada
    if(counter > 0):
        #Dividimos cada string por las comas como elemento separador
        values = line.strip().split(",")
        #A√±adimos cada valor a su respectiva columna del diccionario
        for i in range(len(col_names)):
            main_dict[col_names[i]].append(values[i])
    counter += 1

print("El data set tiene %d filas y %d columnas"%(counter-1, n_cols))

#Convertimos el diccionario procesado a Data Frame y comprobamos que los datos son correctos
df7 = pd.DataFrame(main_dict)
print(df7.head())

#Elegimos donde guardarlo (en la carpeta athletes es donde tiene m√°s sentido por el contexto del an√°lisis)
filename5 = "athletes/downloaded_medals_irene."
fullpath5 = mainpath + "/" + filename5

#Lo guardamos en CSV, en JSON o en Excel seg√∫n queramos
df7.to_csv(fullpath+"csv")
df7.to_json(fullpath+"json")
df7.to_excel(fullpath+"xls")
print("Los ficheros se han guardado correctamente en: " + fullpath5)
```

### 4. Creamos una funci√≥n

Empleamos la l√≥gica del Apartado 2 para crear una funci√≥n, que obteniendo unos par√°metros genera una salida. 

```python
def downloadFromURL(url, filename, sep, delim, encoding, mainpath): #Para definir una funci√≥n y sus par√°metros de entrada

    #Identamos todo para indicar que se haya dentro de la funci√≥n
    import urllib3
    http = urllib3.PoolManager()
    r = http.request('GET', url)
    print("El estado de la respuesta es %d" %(r.status))
    response = r.data

    #El objeto response contiene un string binario, as√≠ que lo convertimos a un string descodific√°ndolo en UTF-8
    str_data = response.decode(encoding)    
    print("El objeto str_data es:")
    print(str_data)
    print(type(str_data))

    #Dividimos el string en un array de filas, separ√°ndolo por intros
    lines = str_data.split(delim)
    print("El objeto lines es:")
    print(lines)
    print(type(lines))
    
    #La primera l√≠nea contiene la cabecera, as√≠ que la extraemos
    cols = lines[0].split(sep)
    n_cols = len(cols)
    print(cols)
    
    #Generamos un diccionario vac√≠o donde ir√° la informaci√≥n procesada desde la URL externa
    counter = 0
    main_dict = {}
    for col in cols:
        main_dict[col] = []
        
    #Procesamos fila a fila la informaci√≥n para ir rellenando el diccionario con los datos como hicimos antes
    for line in lines:
        #Nos saltamos la primera l√≠nea que es la que contiene la cabecera y ya tenemos procesada
        if(counter > 0):
            #Dividimos cada string por las comas como elemento separador
            values = line.strip().split(sep)
            #A√±adimos cada valor a su respectiva columna del diccionario
            for i in range(len(cols)):
                main_dict[cols[i]].append(values[i])
        counter += 1
    
    print(cols, n_cols)
    print("El data set tiene %d filas y %d columnas"%(counter-1, n_cols))
    
    #Convertimos el diccionario procesado a Data Frame y comprobamos que los datos son correctos
    df9 = pd.DataFrame(main_dict)
    print(df9.head())
    
    #Elegimos donde guardarlo (en la carpeta athletes es donde tiene m√°s sentido por el contexto del an√°lisis)
    fullpath5 = mainpath + "/" + filename5

    #Lo guardamos en CSV, en JSON o en Excel seg√∫n queramos
    df9.to_csv(fullpath+".csv")
    df9.to_json(fullpath+".json")
    df9.to_excel(fullpath+".xls")
    print("Los ficheros se han guardado correctamente en: "+ fullpath)
    
    #Para indicar que la funci√≥n ha acabado hay que poner un return
    return df9
```
