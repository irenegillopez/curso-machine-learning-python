# Máquinas vectoriales - Fronteras no lineales

# 0. Preparación previa

```python
import matplotlib.pyplot as plt
import numpy as np

from sklearn.datasets.samples_generator import make_circles, make_blobs #generador de muestras en círculos
from sklearn.svm import SVC

from mpl_toolkits import mplot3d

from ipywidgets import interact, fixed 
```

```python
X, Y = make_circles(100, factor = .1, noise = .1) #Generamos dos vectores X, Y con 100 observaciones
```

```python
#ax = eje, plot_support=pintar los pasillos
#Función para hacer la representación gráfica

def plt_svc(model, ax=None, plot_support=True):
    """Plot de la función de decisión para una clasificación en 2D con SVC""" #Descripción de la función

    #comprueba si hay un dibujo ya creado
    if ax is None: #si no
        ax = plt.gca() #creamos el dibujo
    
    #límites del gráfico
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    ##Generamos la parrila de puntos para evaluar el modelo
    xx = np.linspace(xlim[0], xlim[1], 30)
    yy = np.linspace(ylim[0], ylim[1], 30)
    Y, X = np.meshgrid(yy,xx) #juntamos xx e yy
    
    xy = np.vstack([X.ravel(), Y.ravel()]).T #la T es de tupla
    
    #Aplicamos la función de decisión a estos puntos
    P = model.decision_function(xy).reshape(X.shape)
    
    ##Representamos las fronteras y los márgenes del SVC, k = negro
    ax.contour(X,Y,P, colors="k", levels=[-1,0,1], alpha = 0.5, linestyles=["--", "-", "--"])
    
    print(model.support_vectors_) #los puntos que quedan en el vector soporte
    
    if plot_support:
        ax.scatter(model.support_vectors_[:,0], 
                   model.support_vectors_[:,1], 
                   s=300, linewidth=1, facecolors = "blue");
    
    
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
```

```python
model = SVC(kernel="linear")
model.fit(X, Y)

plt.scatter(X[:,0], X[:,1], c = Y, s = 50, cmap = "autumn")
plt_svc(model, plot_support=False)
```

Evidentemente estos hiperplanos lineales no pueden dividir correctamente el dataset. Una solución sería considerar que realmente, en una dimensión 3D, los puntos amarillos están por encima de los puntos rojos. Por lo que se podría hacer un hiperplano separador entre ambas plantas. Proyección en z.

Para ello, elevamos la dimensión, establecemos un separador aleatorio heights:

```python
heights = np.exp(-(X**2).sum(1)) #Creamos un separador aleatorio con forma de Campana de Gauss
heights #Corresponde a las alturas de cada punto
```

```python
def plot_3D(elev=30, azim=30, X=X, Y=Y, heights=heights): #elevación en grados, componente azimutal
    ax = plt.subplot(projection="3d")
    ax.scatter3D(X[:,0], X[:,1], heights, c=Y, s= 50, cmap="autumn")
    ax.view_init(elev=elev, azim=azim)
    
    ax.set_xlabel("x[0]")
    ax.set_ylabel("y[1]")
    ax.set_zlabel("heights")
```

```python
interact(plot_3D, elev=[-90,-60,-30,0,30,60,90], 
         azim=[-180,-150,-120,-90,-60,-30,0,30,60,90,120,150, 180], 
         X = fixed(X), Y = fixed(Y), heights = fixed(heights))
```

# Otra opción

En lugar de aplicar el kernel linear vamos a aplicar el radial basis function (rbf)
El kernel es radial

```python
rbf = SVC(kernel="rbf", C=1E6)
rbf.fit(X,Y)

plt.scatter(X[:,0], X[:,1], c = Y, s = 50, cmap = "autumn")
plt_svc(rbf)
```

# X. Si no tuviesemos tan clara la separación -> Ajustar los parámetros del SVM

Para ello generamos un dataset random, pero con una desviación estándar mayor a las trabajadas hasta ahora, como podemos ver en la imagen, aquí es mucho más difícil crear un hiperplano recto

```python
X, Y = make_blobs(n_samples = 100, centers = 2, random_state = 0, cluster_std = 0.9)

plt.scatter(X[:,0], X[:,1], c = Y, s = 50, cmap = "autumn")
```

La función de SVM tiene una opción que permite "relajar" las restricciones del margen, a expensas de que quede un poco mal clasificado. Para ello modificamos el valor de C, para valores muy altos de C, el margen es muy estricto.



```python
#Hacemos dos figuras para diferentes valores de C

fig, ax = plt.subplots(1,2, figsize=(16,6)) #Creamos 2 figuras, 1 fila, 2 columnas
fig.subplots_adjust(left = 0.05, right = 0.95, wspace=0.1) #Ajustamos los márgenes

for ax_i, C in zip(ax, [10.0, 0.1]):
    model = SVC(kernel="linear", C=C)
    model.fit(X,Y)
    ax_i.scatter(X[:,0],X[:,1], c = Y, s = 50, cmap="autumn")
    plt_svc(model, ax_i)
    ax_i.set_title("C = {0:.1f}".format(C), size = 15)
```

El corredor es más estrecho cuando C es más alto.
