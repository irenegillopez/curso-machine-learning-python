### Borrado de valores que faltan (falta el 1-3%)

```python
df2 = df1  #Copia para no borrar todo con el ejemplo

#Caso menos drástico (All): borramos sólo si todos los valores son NaNs
df2.dropna(axis=0,how="all") #Borramos filas (eje X)

#Caso más drástico (Any): borramos sólo si 1 valor tiene NaN
df2.dropna(axis=1,how="any") # Borramos columnas (eje Y)
```

### Inferencia o cómputo de valores que faltan

Consiste en reemplazar los valores que faltan por otros valores -> 0, un string, el promedio del resto de valores... depende del contexto de la información.

```python
df3 = df1 #Vamos a sobreescribir en un df nuevo

df3 = df3.fillna(0) #Cambiar por 0
df3 = df3.fillna("Desconocido") #Cambiar por String

# Reemplazar dependiendo de la columna:

df3["body"] = df3["body"].fillna(0) #Porque el campo es numérico
df3["home.dest"] = df3["home.dest"].fillna("Desconocido") #Porque el campo es un string

df3["age"] = df3["age"].fillna(df3["age"].mean()) #Por el promedio
df3["age"] = df3["age"].fillna(method="ffill") #Con el valor inmediatamente posterior ffill  = forward
df3["age"] = df3["age"].fillna(method="backfill") #Con el valor inmediatamente anterior backfill = bakward 
```

### Variables dummy 

Variables artificiales, para las variables categóricas -> creamos dos variables dummy para la columna "sex" -> 
columna male y columna female. Crear tantas columnas como categorías tenga.

```python
df4 = df1 #Vamos a sobreescribir en un df nuevo

dummy_sex = pd.get_dummies(df4["sex"], prefix="sex")  #De la columna Sex, crea columnas con prefijo sex.
dummy_sex.head()
```

Eliminar la variable original, categórica, y añadir las nuevas columnas a la tabla.

```python
df4 = df4.drop(["sex"], axis=1)
df4
```

```python
column_name = df4.columns.values.tolist()
column_name
```

Para añadir (concatenar) las columnas con las variables dummy

```python
df5 = pd.concat([df4, dummy_sex], axis=1)
```

```python
# Función para crear dummies:

def createDummies(df, var_name):
    dummy = pd.get_dummies(df[var_name], prefix=var_name)
    df = df.drop(var_name, axis=1)
    df = pd.concat([df, dummy], axis=1)
    return df
```

```python
df6 = df1 

createDummies(df6, "sex")
```
