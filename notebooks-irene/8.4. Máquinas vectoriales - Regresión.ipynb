# SVM para Regresión

Queremos un SVM que se ajuste lo mayor posible a los datos

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Generamos un dataset aleatorio

X = np.sort(5*np.random.rand(200,1),axis=0) #empiezan en 0 y acaban en 5
Y = np.sin(X).ravel() #usamos la función seno ya que no se puede ajustar por ningún tipo de recta
Y[::5] += 3*(0.5 - np.random.rand(40)) #cada 5 puntos ponemos un outlier
```

```python
#Representación gráfica

plt.scatter(X,Y, color="darkorange", label="data")
```

```python
from sklearn.svm import SVR
```

```python
C=1e3

#Creamos diferentes svr's con diferentes kernel's

svr_lin = SVR(kernel="linear", C=C)
svr_rbf = SVR(kernel="rbf", C=C, gamma=0.1)
svr_pol = SVR(kernel="poly", C=C, degree=3)
```

```python
#creamos 3 líneas de regresión en función del kernel creando el modelo y usando el predict

y_lin = svr_lin.fit(X,Y).predict(X)
y_rbf = svr_rbf.fit(X,Y).predict(X)
y_pol = svr_pol.fit(X,Y).predict(X)
```

```python
lw = 2 #linewidth
plt.figure(figsize=(16,9)) #tamaño figura
plt.scatter(X,Y,color="darkorange", label ="data") #pintamos los puntos
plt.plot(X,y_lin, color="navy", lw = lw, label = "SVM Lineal") #línea linear
plt.plot(X,y_rbf, color="c", lw=lw, label="SVM Radial") #línea radial
plt.plot(X,y_pol, color="red", label="SVM Polinómico") #línea polinómico
plt.xlabel("x")
plt.ylabel("y")
plt.title("Support Vector Regression")
plt.legend() #leyenda
plt.show()

#la que mejor predice es la radial, se ajusta mucho a los puntos independientemente del ruido
```

Las SVM son más útiles cuando hay muuuchos rasgos
